```python
import pandas as pd
```

```python
coop1 = pd.read_csv("Part2 Data/Canada/Co-operation on innovation activities, by type and location of co-operation partners by industry and enterprise size.csv")
```

```python
def find_coop(dataset, col):
    filtered_sector = dataset[
        dataset['North American Industry Classification System (NAICS)'].isin([
            'Information and cultural industries [51]',
            'Professional, scientific and technical services [54]',
            'Manufacturing [31-33]'
        ])
    ]

    selected_columns = filtered_sector[['North American Industry Classification System (NAICS)', 
                                    col, 
                                    'STATUS']]
    return selected_columns
```

```python
find_coop(coop1,"Co-operation on innovation activities")
```

```python
coop2 = pd.read_csv("Part2 Data/Canada/Most critical co-operation partner for the business_s innovation activities, by industry and enterprise size.csv")
find_coop(coop2, "Most critical type of co-operation partner")
```

```python
coop2 = pd.read_csv("Part2 Data/Canada/Most critical co-operation partner for the business_s innovation activities, by industry and enterprise size.csv")
find_coop(coop2, "Most critical type of co-operation partner")
```
